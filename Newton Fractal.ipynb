```python
from math import sin, tan
import matplotlib.pyplot as plt
import numpy as np
a= np.linspace(-75, 50,50)
b= np.linspace(-75,50,50)
for i in range (0,50):
    for j in range (0, 50):
        c=a[i]+1j*b[j]
        z=c
        f=2*(z**5)-3*(z**4)+2*(z**3)-3*(z**2)-144*z+216
        der=10*(z**4)-12*(z**3)+6*(z**2)-6*z-144
        zn=z-f/der
        while abs(zn-z)>0.0000001:
            z=zn
            zn=zn-(2*(z**5)-3*(z**4)+2*(z**3)-3*(z**2)-144*z+216)/(10*(z**4)-12*(z**3)+6*(z**2)-6*z-144)
        #print(zn)
        if abs((zn.real+zn.imag)-(3))==0:
            plt.scatter(c.real, c.imag, color="darkgreen")
        elif abs((zn.real+zn.imag)-(3))<0.1:
            plt.scatter(c.real, c.imag, color="chartreuse")
        elif abs((zn.real+zn.imag)-(-3))==0:
            plt.scatter(c.real, c.imag, color="navy")
        elif abs((zn.real+zn.imag)-(-3))<0.1:
            plt.scatter(c.real, c.imag, color="blue")
        else:
            plt.scatter(c.real, c.imag, color="maroon")
#print('The zero of the function is: ',zn)
#plt.rcParams['figure.figsize']=[10, 10]
plt.show()
```
